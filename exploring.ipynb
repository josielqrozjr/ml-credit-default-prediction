### Descrição da base

- Total de colunas


```python
schema = conn.execute(f"""
DESCRIBE
SELECT DISTINCT(*)
FROM '{all_data}'; -- Pode ajustar ou não o caminho para um dos arquivos Parquet gerados
""").fetchdf()

print(schema)
```
